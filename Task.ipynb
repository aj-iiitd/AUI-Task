```
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

**Question 1**

```
print("Enter the word to find hypernym :")
word = input()
```

```
syn = wordnet.synsets(word)[0] 
print(word,"->",end=" ")
while(syn.hypernyms()[0].name()!=syn.root_hypernyms()[0].name()):
  print(syn.hypernyms()[0].name().split('.')[0],"->",end=" ")
  syn = syn.hypernyms()[0]
print(syn.hypernyms()[0].name().split('.')[0])
```
